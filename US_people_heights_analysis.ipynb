```python
import pandas as pd
import numpy as np
import statistics as st
import statsmodels as stats
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Problem statement first
- All calculations are in Centimeters.
- Data generated is synthetic. 

#### U.S. Mens attributes

```python
m_mean = 175.5
m_std = 7.4
```

```python
#Generating a population data bases on mean and standard deviation parameters using a numpy liberary.
np.random.seed(42)
m = np.random.normal(m_mean,m_std, 1000)
```

```python
sns.histplot(m, bins = 100)
plt.title("Normally distributed heights of 1000 mens in the US")
```

```python
# min 2.2 percent height of tallest persons of population.

m_mean + 2 * m_std 
```

#### U.S. Womens attributes 

```python
w_mean = 161.8
w_std = 6.9
```

```python
#Generating a population data bases on mean and standard deviation parameters using a numpy liberary.
np.random.seed(42)
w = np.random.normal(w_mean, w_std, 1000)
```

```python
sns.histplot(w, bins = 100)
plt.title("Normally distributed heights of 1000 women in the US")
```

```python
# min 2.2 percent height of tallest persons of population.

w_mean + 2 * w_std
```

#### Comparision of both men and women distributions in single graph

```python
sns.histplot(m, bins = 80, alpha = 0.7, label = "m")
sns.histplot(w, bins = 80, alpha = 0.5, label = "w")
plt.legend(loc = 'upper right')
```

# Conclusion: 
- Based on the above graph and statistical findings we can conclude that average mens height ranges in an interval of 175 to 180 cm, while womens height ranges in 157 to 165 cm.
- The heights of men and women having common height range interval is 160 to 175 cm.
- The right outlier of womens height appears to be around mean of mens height, while left outlier of mens height apperars to be around mean of womens height.
